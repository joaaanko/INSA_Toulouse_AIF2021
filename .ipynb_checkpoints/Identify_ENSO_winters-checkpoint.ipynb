```python
%matplotlib inline

import os
os.environ["OMP_NUM_THREADS"] = "1"
import numpy as np
import xarray as xr
from scipy import signal
from netCDF4 import Dataset

from cartopy import config
import cartopy.feature as cfeature
import cartopy.crs as ccrs
from cartopy.mpl.geoaxes import GeoAxes
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import AxesGrid
from mpl_toolkits.axes_grid1 import ImageGrid

from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import pandas as pd
from eofs.standard import Eof
```

## Opening files

Path to files

```python
# datapath="/Data/skd/scratch/cmi022/LargeEnsembles_CLIVAR/CESM_LENS/"
datapath = os.path.join('.','Data_CESM1-CAM5')
plots=os.path.join('.','Plots')
```

Opens files

```python
#Monthly SST over 192001-210012
filenc = 'ts_Amon_CESM1-CAM5_historical_rcp85_r11i1p1_192001-210012.nc'
f = xr.open_dataset(os.path.join(datapath,filenc),mask_and_scale=True,decode_times=True)

lat = f['lat']
lon = f['lon']
time = f['time']
sst = f['ts']
f.close()
del f
```

## SST monthly anomalies

Splits files into 40-y periods shifted by 1y. <br>
Removes the seasonal cycle from SST. <br>
Detrends the signal. <br>
Calculates SST anomalies. <br>


IMPORTANT : here I choose p = 2 for the sake of the example. EOF1 is negative over the NINO3.4 domain for this period of 40 years and that is what we want to correct.

```python
# Number of months in one 40-yr period
nmon=12*40

# First index of the periods
a=np.arange(0,len(time)+1,12)

#Loop over the periods
#for p in range(len(time)/12):
#for p in range(181):
p=0
sst_40=sst[a[p]:a[p]+nmon,:,:]
print('***** SST ******')
time_40=time[a[p]:a[p]+nmon]


#Select the months and average:
seascyc=np.zeros((12,len(lat),len(lon)))

for m in range(12):
    mon=int(m+1)
    i=np.where(time_40.dt.month==mon)[0]
    seascyc[m,:,:]=np.mean(sst_40[i,:,:],axis=0)
    del i
print('***** Seasonal cycle *****') 
    

# Calculate the anomalies
ssta_40=np.zeros(np.shape(sst_40))

for i in range(nmon):
    mon=(time_40.dt.month-1)[i]
    ssta_40[i,:,:]=sst_40[i,:,:]-seascyc[mon,:,:]
print('***** SST anomalies *****')

del sst_40
del seascyc
# Remove the trend
sstdet_40=np.zeros(np.shape(ssta_40))

for lo in range(len(lon)):
    for la in range(len(lat)):
        sstdet_40[:,la,lo] = signal.detrend(ssta_40[:,la,lo])

del ssta_40
print('***** Detrended SST anomalies*****')
```

```python
StartY=np.array(time_40.dt.year[0])
EndY=np.array(time_40.dt.year[-1])
```

## Empirical Orthogonal Function

Select the North Pacific : Domain 15째S-15째N / 140째E-80째W.

```python
# Select SST where -15 <= lat <= 15 and 140 <= lon <=280

#Index of the latitudes and longitudes

ilatmin = np.where(np.round(lat) == -15)[0]
ilatmax = np.where(np.round(lat) == 15)[0]
ilonmin = np.where(np.round(lon) == 140)[0]
ilonmax = np.where(np.round(lon) == 280)[0]

ssta_40_pac = sstdet_40[:,int(ilatmin):int(ilatmax)+1,int(ilonmin):int(ilonmax)+1]
lat_pac = lat[int(ilatmin):int(ilatmax)+1]
lon_pac = lon[int(ilonmin):int(ilonmax)+1]
print(np.shape(ssta_40_pac))
```

Square-root of cosine of latitude weights are applied before the computation of EOFs.

```python
coslat = np.cos(np.deg2rad(np.array(lat_pac)))
wgts = np.sqrt(coslat)[...,np.newaxis]

print(np.shape(wgts))
```

PCA.

```python
solver = Eof(ssta_40_pac, weights = wgts)
#solver = Eof(ssta_40_pac)
eigenvalues = solver.eigenvalues()
total_variance = solver.totalAnomalyVariance()
varfrac = solver.varianceFraction()
eofs = solver.eofsAsCovariance()
#eofs = solver.eofs(neofs=2)
#eofs = solver.eofsAsCorrelation()
pcs = solver.pcs()
#pcs = solver.pcs(npcs = 2, pcscaling = 0)
#pcs_norm = solver.pcs(pcscaling = 1)
```

Plot of the 3 first spatial structures (EOFs)

```python
def plot_background(ax):
    ax.gridlines(crs=ccrs.PlateCarree(), linewidth=0.5, color='gray', alpha=0.5, linestyle='-')
    ax.set_xticks(np.arange(140,290,20),crs=ccrs.PlateCarree())
    ax.set_yticks(np.arange(-15,15,5.), crs=ccrs.PlateCarree())
    lon_formatter = LongitudeFormatter(zero_direction_label=True)
    lat_formatter = LatitudeFormatter()
    ax.xaxis.set_major_formatter(lon_formatter)
    ax.yaxis.set_major_formatter(lat_formatter)
    ax.coastlines('50m')
    
    return ax
```

```python
axes_class = (GeoAxes, dict(map_projection = ccrs.PlateCarree(central_longitude = 210)))
fig = plt.figure(figsize=(20,10))
fig.suptitle('EOFs : detrended SST anomalies '+str(StartY)+'-'+str(EndY), fontsize=16)


axgr = AxesGrid(fig,111, axes_class = axes_class,
        nrows_ncols=(3,1),
        axes_pad=0.6,
        cbar_location='right',
        cbar_mode='single',
        cbar_pad=0.3,
        cbar_size='1%',
        label_mode='')
clevs = np.linspace(-2,2,21)

for i, ax in enumerate(axgr):
    ax = plot_background(ax)
    ax.set_title('EOF'+str(i+1)+' ('+str(int(varfrac[i]*100))+'%)', fontsize=10,loc='center')
    cf = ax.contourf(lon_pac,lat_pac,eofs[i],clevs,transform=ccrs.PlateCarree(),cmap='RdBu_r')
    c = ax.contour(lon_pac,lat_pac,eofs[i],levels=clevs,colors='black',linewidths=0.5,transform=ccrs.PlateCarree())
    axgr.cbar_axes[i].colorbar(cf)
    ax.clabel(c,clevs,inline=True,fontsize=10)

plt.show()


#figname=os.path.join(plots,'eofs_sst_pacn_'+str(StartY)+'-'+str(EndY))
#fig.savefig(figname+'.pdf')
#fig.savefig(figname+'.png')
```

Note that EOF1 is mainly negative for this period when it should be positive. <br>
In the following section, we inverse the sign of EOF1 and pc1 to make it positive. <br>

## Correct the sign of EOF1 and PC1

Select NINO3.4 domain (5N-5S, 170W-120W/190E-240E)

```python
ilatmin = np.where(np.round(lat_pac) == -5)[0]
ilatmax = np.where(np.round(lat_pac) == 5)[0]
ilonmin = np.where(np.round(lon_pac) == 190)[0]
ilonmax = np.where(np.round(lon_pac) == 240)[0]

ssta_40_nino34 = ssta_40_pac[:,int(ilatmin):int(ilatmax)+1,int(ilonmin):int(ilonmax)+1]
lat_nino34 = lat_pac[int(ilatmin):int(ilatmax)+1]
lon_nino34 = lon_pac[int(ilonmin):int(ilonmax)+1]


eof1_nino34 = np.mean(eofs[0,int(ilatmin):int(ilatmax),int(ilonmin):int(ilonmax)])
```

Check if EOF1 > 0 over the Nino3.4 domain. <br>
If not, inverse the sign of EOF1 and PC1. <br>

```python
if eof1_nino34 <0 : #eof1_nino34 < 0 : the EOF1 is negative over the NINO3.4 domain.
    eofs[0]=eofs[0]*(-1) # multiply by (-1) to inverse the sign
    pcs[:,0]=pcs[:,0]*(-1)
```

Plot the EOF1 multiplied by (-1).

```python
axes_class = (GeoAxes, dict(map_projection = ccrs.PlateCarree(central_longitude = 210)))
fig = plt.figure(figsize=(20,10))
fig.suptitle('EOFs : detrended SST anomalies '+str(StartY)+'-'+str(EndY), fontsize=16)


axgr = AxesGrid(fig,111, axes_class = axes_class,
        nrows_ncols=(3,1),
        axes_pad=0.6,
        cbar_location='right',
        cbar_mode='single',
        cbar_pad=0.3,
        cbar_size='1%',
        label_mode='')
clevs = np.linspace(-2,2,21)

for i, ax in enumerate(axgr):
    ax = plot_background(ax)
    ax.set_title('EOF'+str(i+1)+' ('+str(int(varfrac[i]*100))+'%)', fontsize=10,loc='center')
    cf = ax.contourf(lon_pac,lat_pac,eofs[i],clevs,transform=ccrs.PlateCarree(),cmap='RdBu_r')
    c = ax.contour(lon_pac,lat_pac,eofs[i],levels=clevs,colors='black',linewidths=0.5,transform=ccrs.PlateCarree())
    axgr.cbar_axes[i].colorbar(cf)
    ax.clabel(c,clevs,inline=True,fontsize=10)

plt.show()


#figname=os.path.join(plots,'eofs_sst_pacn_'+str(StartY)+'-'+str(EndY))
#fig.savefig(figname+'.pdf')
#fig.savefig(figname+'.png')
```

## Standardised 3-month running averaged PC1

3-month running averaged PC1

Using convolution.

```python
pc_3mavg = np.zeros(len(pcs[:,0]))

pc_3mavg[0] = np.mean(pcs[0:2,0]) #average on the two first months : JF
pc_3mavg[-1] = np.mean(pcs[-2:,0]) #average on the two last months : ND

pc_3mavg[1:-1] = np.convolve(pcs[:,0], np.ones((3,))/3, mode='valid')
```

Standardize 3-month running averaged PC1

```python
PC1_standardised =np.zeros(480)
mean = np.mean(pc_3mavg[:])
std = np.std(pc_3mavg[:]) 

PC1_standardised[:] = (pc_3mavg[:] - mean)/std
```

```python
fig = plt.figure(figsize=(20,5))
ticks_label = np.arange(int(StartY),int(EndY)+1,1)
fig.suptitle('Standardised 3-month running averaged PC1 : '+ str(StartY)+'-'+str(EndY), fontsize=16)
ax = fig.add_subplot(111)
ax.set_xticks(np.arange(0,481,12))
ax.set_xticklabels(ticks_label)
plt.ylabel('PC1')
plt.xlabel('Months')
plt.axhline(0, color='k')
plt.axhline(1, color='red', linewidth=0.5, linestyle='--')
plt.axhline(-1, color='blue', linewidth=0.5, linestyle='--')
plt.ylim(-3, 3)
colormat=np.where(PC1_standardised>0, 'red','blue')
plt.bar(np.arange(0,480), PC1_standardised, width=0.5, color=colormat)

plt.show()

figname=os.path.join('.','pc3mstd_sst_pacn_'+str(StartY)+'-'+str(EndY))
fig.savefig(figname+'.pdf')
fig.savefig(figname+'.png')
```

## TEXT Files

Identify winters (DJF) for which the standardised 3-month running averaged first principal component is >1 (El Ni챰o case) or <-1 (La Ni챰a case).

```python
#Index in list PC1 : PC1 > 1
index_nino = np.where(PC1_standardised > 1)[0] #also corresponds to the indexes of December in time_40 (192001-195912)
index_nina = np.where(PC1_standardised < -1)[0]
el_nino_D, la_nina_D = [], []
for i in range(1,len(index_nino)-2) :
    #Checking if PC1 > 1 for OND NDJ DJF and JFM 
    if time_40.dt.month[index_nino[i]]==12 and time_40.dt.month[index_nino[i+1]]==1 and time_40.dt.month[index_nino[i+2]]==2:
        el_nino_D.append(index_nino[i])

for j in range(1,len(index_nina)-2):
    #Checking if PC1 < -1 for OND NDJ DJF and JFM
    if time_40.dt.month[index_nina[j]]==12 and time_40.dt.month[index_nina[j+1]]==1 and time_40.dt.month[index_nina[j+2]]==2:
        la_nina_D.append(index_nina[j])

print(el_nino_D)

print(la_nina_D)
```

```python
file_nino = "NINO_CESM1-CAM5.txt" #1 text file for El Ni챰o events for the member CESM1-CAM5
file_nina = "NINA_CESM1-CAM5.txt" #1 text file for La Ni챰a events for the member CESM1-CAM5

if os.path.isfile(file_nino) or os.path.isfile(file_nina):
    mode ="a"
else :
    mode="w"

f = open(file_nino,mode)
for index in el_nino_D : #Indexes of 'El Ni챰o Decembers' for 1 period of 40 years
    f.write(str(index)+' ')
f.write('\n') #End of line : new line = new period of 40 years
f.close()

f = open(file_nina,mode)
for index in la_nina_D : #Indexes of 'La Ni챰o Decembers' for 1 period of 40years
    f.write(str(index)+' ')
f.write('\n') #End of line : new line = new period of 40 years
f.close()
```
